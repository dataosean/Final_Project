```python
# 자료형 변환
info['Review_score'] = info['Review_score'].astype(float)
info['Review_counts'] = info['Review_counts'].astype(int)
```

```python
# 띄어쓰기, 개행 제거
info['Review_text'] = info['Review_text'].str.replace('\r\n', " ")
info['Review_text'] = info['Review_text'].str.replace('\n', " ")
```
